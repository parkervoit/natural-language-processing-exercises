```python
def get_blog_articles(url_list):
    dict_list = []
    for i in url_list:
        headers = {'User-Agent': 'Codeup Data Science'}
        response = get(i, headers=headers)
        soup = BeautifulSoup(response.text)
        text = soup.get_text(strip = True)
        title =  soup.title.string
        dictionary = {'title':title,
                      'text': text}
        dict_list.append(dictionary)
    return pd.DataFrame(dict_list)
```

```python
get_blog_articles(url_list)
```

```python
def get_blog_articles(url_list, body):   
    dict_list = []
    for i in url_list:
            headers = {'User-Agent': 'Codeup Data Science'}
            response = get(i, headers=headers)
            soup = BeautifulSoup(response.text)
            text = soup.find('div', class_ = body).get_text(strip = True)
            title =  soup.title.string
            dictionary = {'title':title,
                          'text': text}
            dict_list.append(dictionary)
    return dict_list
```

```python
body = 'jupiterx-post-content'
get_blog_articles(url_list, body)
```
